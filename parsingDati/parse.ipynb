```python
import pandas as pd
```

```python
# Data una porzione del dataframe che parte dalla riga con la scritta blue estrae un altro 
# dataframe realtivo agli spostamenti di una singola biblioteca
def estraiDati(singola):
    biblioteche_prestanti = singola.iloc[2:, 0][singola.iloc[2:, 0:6].iloc[:, 0].notna()].values
    prestiti_per_biblioteca = singola.iloc[2:, 2][singola.iloc[2:, 0:6].iloc[:, 2].notna()].values
    biblioteca_richiedente = [singola.iloc[0][0]] * len(biblioteche_prestanti)
    data = ["12-09-2022"] * len(biblioteche_prestanti)

    temp = pd.DataFrame(list(zip(data, biblioteca_richiedente, biblioteche_prestanti, prestiti_per_biblioteca)),
               columns =['Data','BibliotecaRichiedente', 'BibliotecaPrestante','NumeroLibri'])
    
    return temp
```

```python
def splitDataframe(filename):
    df = pd.read_csv(filename, skiprows=19) #salta le righe di output del gestionale

    # Raggruppa in modo da inserire in result tutti i dataframe delle singole biblioteche, da passare poi in estraiDati()
    grp = (df.isnull().sum(axis=1) == df.shape[1]).cumsum()
    gr = df.dropna(axis=0, thresh=1).groupby(grp)
    result = [ gr.get_group(key) for key in gr.groups ]

    print("Biblioteche", len(result))

    return result
```

```python
tutte = splitDataframe('StatisticaServizio.csv')
final = pd.DataFrame(columns =['Data','BibliotecaRichiedente','BibliotecaPrestante','NumeroLibri'])

for single in tutte:
    final = final.append(estraiDati(single))

final.to_csv("out.csv")
```
